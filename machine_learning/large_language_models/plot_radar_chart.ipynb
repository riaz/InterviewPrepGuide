```python
import matplotlib.pyplot as plt
import numpy as np

# HELM Evaluation results (for example purposes)
labels = ['Writing', 'Roleplay', 'Reasoning', 'Math', 'Coding', 'Extraction', 'STEM', 'Humanities']
zephyr_7B = [7, 6, 8, 7, 5, 6, 7, 7]
mistral_7B = [8, 7, 9, 7, 6, 6, 8, 7]
llama_2_70b = [7, 5, 8, 6, 5, 5, 6, 7]

# Set data
data = [zephyr_7B, mistral_7B, llama_2_70b]
colors = ['b', 'r', 'g']

# Create a radar chart
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

# Plot data
for d, color in zip(data, colors):
    angles = np.linspace(0, 2 * np.pi, len(labels), endpoint=False).tolist()
    #d += d[:1]  # repeat the first value to close the circular graph
    #angles += angles[:1]
    ax.fill(np.concatenate((angles, [angles[0]])), np.concatenate((d, [d[0]])), alpha=0.25, color=color)
    ax.plot(np.concatenate((angles, [angles[0]])), np.concatenate((d, [d[0]])), color=color, linewidth=2)
    #ax.fill(angles, d, alpha=0.25, color=color)
    #ax.plot(angles, d, color=color, linewidth=2)

# Add a title
ax.set_title("HELM Evaluation Results", size=20, color="black", y=1.1)

# Set the labels (around the edge)
ax.set_thetagrids(np.degrees(angles), labels)

plt.legend(['Zephyr 7B Alpha', 'Mistral 7B Instruct', 'Llama-2 70b Chat'], loc='upper right')

plt.show()
```
